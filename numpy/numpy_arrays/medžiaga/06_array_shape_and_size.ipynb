```python
import numpy as np
```

## Shape

### One-dimensional

```python
first_arr = np.arange(20)
first_arr
```

```python
np.shape(first_arr)
```

> Gavome tuple, kuris yra one-dimensional su 20 elementų.

### Two-dimensional

```python
second_arr = np.linspace((1,2), (10,20), 10)
second_arr
```

```python
np.shape(second_arr)
```

> Gavome tuple, kuris yra two-dimensional su 10 eilučių ir 2 stulpeliais

### Three-dimensional

```python
third_arr = np.full((2,2,2), 10)
third_arr
```

```python
np.shape(third_arr)
```

> Gavome tuple, kuris yra three-dimensional su 2 matricom, 2 stulpeliais ir 2 eilutėm.

## Size

```python
np.size(first_arr)
```

```python
np.size(second_arr)
```

```python
np.size(third_arr)
```

> Size parodo kiek elementų turi array
